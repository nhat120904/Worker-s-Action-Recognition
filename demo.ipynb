```python
import pandas as pd
import os

root = "Theme2/Theme2"
sheet_names = ["Markers", "Segment Velocity", "Segment Acceleration"]
for part_folder in os.listdir(root):
    part_path = os.path.join(root, part_folder)
    for folder in os.listdir(part_path):
        folder_path = os.path.join(part_path, folder)
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            
            # Read the specific sheets into a dictionary
            sheets_dict = pd.read_excel(file_path, sheet_name=sheet_names)

            # Overwrite the original file with the selected sheets
            with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
                for sheet_name, data in sheets_dict.items():
                    data.to_excel(writer, sheet_name=sheet_name, index=False)

            print(f"Selected sheets saved and replaced in {file_path}")
```

```python
import os
import pandas as pd
import re

# Define sharpness categories
def categorize_sharpness(sharpness):
    if sharpness >= 85:
        return 'Sharp'
    elif 70 <= sharpness < 85:
        return 'Medium'
    else:
        return 'Blunt'

# Function to extract sharpness from filename and categorize it
def extract_and_categorize_sharpness(filename):
    # Extract sharpness value using regex (assumes sharpness is the number before the last dash)
    sharpness_value = int(re.search(r'-([0-9]+)-', filename).group(1))
    return categorize_sharpness(sharpness_value)

# Function to rename the file with the sharpness category
def rename_file_with_category(file_path):
    # Extract filename and extension
    dirname = os.path.dirname(file_path)
    basename = os.path.basename(file_path)
    filename, ext = os.path.splitext(basename)
    
    # Extract and categorize sharpness
    sharpness_category = extract_and_categorize_sharpness(filename)
    
    # Construct the new filename with the sharpness category
    new_filename = re.sub(r'-[0-9]+-', f'-{sharpness_category}-', filename) + ext
    new_file_path = os.path.join(dirname, new_filename)
    
    # Rename the file
    os.rename(file_path, new_file_path)
    return new_file_path

# File path
file_path = 'MVN-J-Boning-64-001.xlsx'  # Replace with your actual file path

# Rename the file based on sharpness category
new_file_path = rename_file_with_category(file_path)
print(f"File renamed to: {new_file_path}")

# Load the renamed Excel file
sheet_names = ["Marker", "Segment Velocity", "Segment Acceleration"]

# Read the specific sheets into a dictionary
sheets_dict = pd.read_excel(new_file_path, sheet_name=sheet_names)

# Overwrite the renamed file with the selected sheets
with pd.ExcelWriter(new_file_path, engine='xlsxwriter') as writer:
    for sheet_name, data in sheets_dict.items():
        data.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"Selected sheets saved and replaced in {new_file_path}")
```
